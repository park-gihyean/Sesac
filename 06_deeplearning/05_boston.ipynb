```python
mean = np.mean(x_train)
std = np.std(x_train)

x_train = (x_train - mean)/std
x_test = (x_test - mean)/std
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train,
                                                  y_train,
                                                  test_size = 0.33,
                                                  random_state = 777)
```

```python
model = Sequential()
model.add(Dense(64,activation='relu',input_shape =(13,)))
model.add(Dense(32,activation='relu'))
# 숫자를 늘렸다가 줄였다 쓰지는 않는다.
# 결과가 잘 나오는 모델을 참고해서 쓰는것이 좋다.

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics =['mae'])
# mae 오차 절댓값

history = model.fit(x_train, y_train, epochs = 300, validation_data=(x_val,y_val))

```

```python
from keras.datasets.boston_housing import load_data
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import KFold

import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = load_data(seed = 777)
```

```python
mean = np.mean(x_train)
std = np.std(x_train)

x_train = (x_train - mean)/std
x_test = (x_test - mean)/std

k = 3

Kfold = KFold(n_splits = k, shuffle = True, random_state = 777)

def get_model():
  model = Sequential()
  model.add(Dense(64,activation='relu',input_shape =(13,)))
  model.add(Dense(32,activation='relu'))
  # 숫자를 늘렸다가 줄였다 쓰지는 않는다.
  # 결과가 잘 나오는 모델을 참고해서 쓰는것이 좋다.

  model.add(Dense(1))

  model.compile(optimizer='adam', loss='mse', metrics =['mae'])
  # mae 오차 절댓값

  return model

mae_list = []

for train_index, val_index in Kfold.split(x_train):
  x_train_fold, x_val_fold = x_train[train_index], x_train[val_index]
  y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

  model = get_model()

  model.fit(x_train_fold, y_train_fold, epochs = 300, validation_data=(x_val_fold,y_val_fold))

  _,test_mae = model.evaluate(x_test,y_test)
  mae_list.append(test_mae)



history = model.fit(x_train, y_train, epochs = 300, validation_data=(x_val,y_val))
```

```python
print(mae_list)
print(np.mean(mae_list))
```
